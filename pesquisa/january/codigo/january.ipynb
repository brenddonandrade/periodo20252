```python
import yfinance as yf
import pandas as pd
```

## Baixando os dados
Tickers analisados:
- KC=F: café;
- SB=F: açucar;
- CL=F: pretrólio bruto;
- PL=F: Platina

```python
tickers = ['KC=F', 'CB=F', 'PL=F', 'CL=F']

string_tickers = '-'.join(tickers)

# para identificar o nome
string_tickers = string_tickers.replace('=', '-')
period = '10y'
string_tickers += ('-' + period)
file_name = 'data-' + string_tickers + '.csv'

data_analyzed = yf.download(tickers, period=period)
data_analyzed.to_csv(file_name)
```
